```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
	return (x+3)**2
def df(x):
	return 2*(x+3)
```

```python
x=2
learning_rate=0.1
num_iteration =100
x_his=[]
y_his=[]
```

```python
for i in range(num_iteration):
	gradient=df(x)
	x_new=x-gradient*learning_rate
	x_his.append(x)
	y_his.append(f(x))
	x=x_new
```

```python
x_val=np.linspace(-10,4,100)
y_val=f(x_val)
```

```python
plt.plot(x_val, y_val)
plt.scatter(x_his, y_his)
```
